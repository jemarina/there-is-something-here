Тема: "Анализ имиграции"

Иммиграция населения — въезд населения одной страны (государства) в другую на временное или постоянное проживание, рассматриваемый по отношению к стране, куда въезжают мигрант.


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('/content/drive/MyDrive/Grazhdanstvo.csv')
df1.head(1405)
```

```python
import pandas as pd
df2 = pd.read_csv('/content/drive/MyDrive/Strana_rozhdenia.csv')
df2.head(1405)
```

Нас интересуют только названия стран, период времени и количество людей,покину поэтому удалим столбцы, которые не входят в наше исследование.

```python
df1.drop([ "LAST UPDATE", "DATAFLOW", "freq", "age", "agedef", "unit", "sex", "OBS_FLAG", "citizen" ], axis = 1, inplace = True)
df1
```

Дня начала рассмотрим 2019 год


```python
df1 = df1[df1["TIME_PERIOD"].apply(str).str.contains("2019")]
```

```python
df1 = df1.sort_values(by = 'OBS_VALUE', ascending=False).head(34)
df1
```

Формируем первые 34 значения

```python
%matplotlib inline
plt.rcParams["figure.figsize"] = (16,5)
df1.pivot_table(values='OBS_VALUE',index ='geo',fill_value=None,margins = False,dropna= True).plot()
plt.ylabel('Количество людей')
plt.xlabel('Страны')
plt.title('График имиграции')
```

По полученному графику можно сделать вывод, что в 2019 году в Чехии был самый высокий уровень имиграции


```python
df2.drop(["LAST UPDATE", "DATAFLOW", "freq", "age", "agedef", "unit", "sex", "OBS_FLAG", "c_birth"], axis = 1, inplace = True)
df2
```

Удаляем лишние столбцы

```python
df2 = df2[df2["TIME_PERIOD"].apply(str).str.contains("2018")]
```

```python
top_25_1 = df2.sort_values(by = 'OBS_VALUE', ascending=False).head(22)
top_25_1
```

```python
df2.plot(x = 'geo', y = "OBS_VALUE", kind = 'bar', rot = 0, fontsize = 5)
plt.rcParams['font.size'] = 20
plt.tick_params(axis = 'both', which = 'major', labelsize = 10)
```

По полученным данным можно сделать вывод, что большинство людей предпочитают гражданство Германии.


```python
df1 = pd.read_csv('/content/drive/MyDrive/Обьеденение.gz')
df1.head(1405)
```

```python
df1.drop(["DATAFLOW", "LAST UPDATE", "freq", "c_birth", "unit","OBS_FLAG"], axis = 1, inplace = True)
df1
```

```python
TR_HEALTH = df1.sort_values(by = 'OBS_VALUE', ascending=False).head(12)
TR_HEALTH
```

Сортируем по убыванию 

```python
%matplotlib inline
plt.rcParams["figure.figsize"] = (20,10)
TR_HEALTH.pivot_table(values='OBS_VALUE',index ='TIME_PERIOD',fill_value=None,margins = False,dropna= True).plot()
plt.xlabel('Год')
plt.title('Рост имиграции')
```

На данном графике мы можем увиидеть, что в 2015 году имиграционный поток был самым высоким.

```python
plt.close()
plt.rcParams["figure.figsize"] = (15,3)
df1.plot(x="TIME_PERIOD", y=["OBS_VALUE", "geo"])

plt.title('Сравнение графиков')
```

Данный график подтверждает выводы, что экономический фактор движение потоков людей зависит от экономического положения в стране.
