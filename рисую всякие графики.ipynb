```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.pie([50, 20,])
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot([6,2, 5, 3, 4], [0, 6, 7, 15, 19],linewidth = 4)
ax.scatter([2, 1, 2, 5, 4], [1, 3, 6, 6, 23], color = 'green')

plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
img = y.reshape(5, 20)

fig, axes = plt.subplots(nrows = 2, ncols = 1)

axes[0].plot(x, y)
axes[1].imshow(img)
plt.show()
```

```python
fig, ax = plt.subplots()
x1 = 10*np.random.rand(200)  
y1 = 10*np.random.rand(200)
ax.scatter(x1, y1)    
plt.show()
```

```python
x = np.zeros(10)

fig, ax = plt.subplots()
ax.plot(x,
        marker = 'o',
        color = 'pink')
ax.plot(x + 1,
        marker = '^',
        markersize = 21,
        color = 'red')
ax.plot(x + 2,
        marker = 's',
        markersize = 10,
        color = 'orange')
ax.plot(x + 3,
        marker = 'X',
        markersize = 20,
        color = 'seagreen')

ax.plot(x + 5,
        marker = '^',
        markersize = 20,
        color = 'grey',
        markerfacecolor = 'black',
        markeredgecolor = 'black',
        markeredgewidth = 3)
ax.plot(x + 6,
        marker = 's',
        markersize = 20,
        color = 'lime',
        linewidth = 4,
        markerfacecolor = 'black',
        markeredgecolor = 'black',
        markeredgewidth = 5)
ax.plot(x + 7,
        marker = 'X',
        markersize = 20,
        color = 'teal',
        markerfacecolor = 'greenyellow',
        markeredgecolor = 'teal',
        markeredgewidth = 3)

fig.set_figwidth(12)
fig.set_figheight(6)
fig.set_facecolor('floralwhite')
ax.set_facecolor('seashell')

plt.show()
```

```python

x = np.random.rand(5000)
y1 = np.random.gamma(1, size = 5000)
y2 = np.random.gamma(3, size = 5000)
y3 = np.random.gamma(8, size = 5000)
y4 = np.random.gamma(4, size = 5000)

fig, ax = plt.subplots()
ax.scatter(x, y1,
           c = [[1, 0, 0, 0.05]])
ax.scatter(x + 1, y2,
           c = [[0.1, 0.63, 0.55, 0.05]])
ax.scatter(x + 2, y3,
           c = '#ad09a305')

ax.scatter(x + 3, y4,
           c = [[0.5, 0, 0.5, 0.05]])
fig.set_figwidth(8)
fig.set_figheight(8)

plt.show()
```

```python
x = np.arange(1, 8)
y = np.random.randint(1, 20, size = 7)

fig, ax = plt.subplots()

ax.bar(x, y)

ax.set_facecolor('white')
fig.set_facecolor('pink')
fig.set_figwidth(12)   
fig.set_figheight(6)   

plt.show()
```

```python
x = np.arange(1, 51)
y = np.random.randint(1, 50, size = 50)

fig, ax = plt.subplots()

bottom_rectangle = np.random.randint(30, size = 50)
ax.bar(x, y, bottom = bottom_rectangle)

fig.set_figwidth(12) 
fig.set_figheight(6) 
fig.set_facecolor('purple')
ax.set_facecolor('floralwhite')

plt.show()
```

```python
x = np.random.rand(500)
y = np.random.rand(500)
fig, axes = plt.subplots(4, 1)
colors = np.random.rand(500)
axes[0].scatter(x, y, c = colors)
axes[0].set_facecolor('purple')
axes[0].set_xticks([])
rgb = np.random.rand(500, 3)
axes[1].scatter(x, y, c = rgb)
axes[1].set_facecolor('red')
axes[1].set_xticks([])
rgba = np.random.rand(500, 4)
axes[2].scatter(x, y, c = rgba)
axes[2].set_facecolor('white')
axes[2].set_xticks([])
col = np.empty_like(y)
below_points = (y < 0.5)   
upper_point = (y >= 0.5)   
col[below_points] = 0
col[upper_point] = 1
axes[3].scatter(x, y, c = col)
axes[3].set_facecolor('pink')
axes[3].set_xticks([])
axes[3].set_yticks([0.0, 0.5, 1.0])
fig.set_figwidth(12)  
fig.set_figheight(12)   
plt.show()
```
